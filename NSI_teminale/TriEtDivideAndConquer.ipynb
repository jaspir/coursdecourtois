# II Diviser pour régner

Pour trier une liste d'entier, on peut la scinder en deux sous-liste de même taille que l'on triera séparement puis ensuite de recombiner les deux listes triées en une seule.


### Exemple :

    soit L=[100, 60, 91, 62, 7, 78, 56, 59, 30, 43]

On divise cette liste en deux listes :

    L1=[100, 60, 91, 62, 7] et  L2=[78, 56, 59, 30, 43]
    
On trie chacune des deux sous-listes :

    L1=[7, 60, 62, 91, 100] et L2=[30, 43, 56, 59, 78]
    
Puis on combine (fusionne) les deux listes:

    L=[7, 30, 43, 56, 59, 60, 62, 78, 91, 100]
    
### C'est la méthode de diviser pour régner

# Définition
"diviser pour régner" ("divide and conquer" en anglais) est une méthode algorithmique consistant à diviser un probleme en sous-probleme du même type ("diviser") . On résout récursivement ou directement chacun des problémes ("régner") puis on recombine les résultats .

Il y a trois phases :
  
+ **Diviser** : on divise le probléme en plusieurs sous-problémes.
+ **Régner** : On résout chacun des sous-problémes
+ **Combiner** : On fusionne les résultats pour arriver à la solution finale.

# IV Recherche d’un élément dans une liste non triée

Le problème est de rechercher la présence d’un élément dans une liste.

En adoptant le paradigme "diviser pour régner", l’idée pour résoudre cette question est de
rechercher récursivement l’élément dans la première moitié de la liste et dans la seconde,
puis de combiner les résultats via l’opérateur logique ``or``.

En effet, l’élément recherché sera dans la liste s’il est dans la première moitié ou dans la
seconde. La condition d’arrêt à la récursivité sera l’obtention d’une liste à un seul élément,
car il est alors immédiat de conclure si l’élément recherché appartient à une telle liste ou
non.

Voici donc les trois étapes de la résolution de ce problème via la méthode "diviser pour
régner" :

+ Rechercher la présence de l’élément dans chacune de ces sous-listes. Arrêter la récursion lorsque les listes n’ont plus qu’un seul élément.
+ Combiner avec l’opérateur logique or les résultats obtenu

# V Recherche d’un élément dans une liste triée

Le problème est de rechercher la présence d’un élément dans une liste préalablement triée.

L’idée pour résoudre cette question est d’utiliser une méthode dichotomique.

La liste étant triée, après comparaison avec l’élément du "milieu" il est en effet facile de voir
dans quelle moitié peut éventuellement se trouver l’élément cherché. On aura plus alors
qu’à recommencer récursivement la recherche.

Le terme dichotomie provient du grec ancien "dikhotomia", signifiant "division en deux parties".
Voici donc les trois étapes de la résolution de ce problème via la méthode "diviser pour
régner".
+ Diviser la liste en deux sous-listes en la “coupant” par la moitié.
+ Rechercher récursivement la présence de l’élément recherché dans la “bonne” des deux
sous-listes après l’avoir comparé à l’élément situé au milieu de la liste.
+ Pas de résultats à combiner puisque l’on ne “travaille” que sur l’une des deux sous-listes.
